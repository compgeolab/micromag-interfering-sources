## Generate dipole moments distribution

```python
def amplitude_lognormal_distribution(size, mean=1E-14, dev_pad=1.5, seed=None):
    """
    Generates amplitude values in a lognormal distribution
    """
    np.random.seed(seed)
    lognormal_distribution = np.random.lognormal(0, dev_pad, size)
    return lognormal_distribution * mean 
```

```python
vector_size1 = 1250
dispersion = np.deg2rad(180)  

alpha = np.deg2rad(np.random.uniform(0,360,vector_size1))
r = np.random.normal(0, dispersion, vector_size1)

x = np.sin(r)*np.cos(alpha)
y = np.sin(r)*np.sin(alpha)
z = np.cos(r)
r_vector1 = np.array([x,y,z])


vector_size2 = 1250
dispersion = np.deg2rad(50)  

alpha = np.deg2rad(np.random.uniform(0,360,vector_size2))
r = np.random.normal(0, dispersion, vector_size2)

x = np.sin(r)*np.cos(alpha)
y = np.sin(r)*np.sin(alpha)
z = np.cos(r)
r_vector2 = np.array([x,y,z])
```

```python
amp1 = amplitude_lognormal_distribution(vector_size1, mean=1.0e-13, dev_pad=1.0)
amp2 = amplitude_lognormal_distribution(vector_size2, mean=1.0e-14, dev_pad=1.5)

# Definindo os bins logarítmicos baseados no intervalo combinado
min_bin = min(amp1.min(), amp2.min())
max_bin = max(amp1.max(), amp2.max())
logbins = np.logspace(np.log10(min_bin), np.log10(max_bin), 100)

plt.figure(figsize=(8, 6))
plt.hist(amp1, bins=logbins, color='blue', alpha=0.6, label='Distribuição 1')
plt.hist(amp2, bins=logbins, color='red', alpha=0.6, label='Distribuição 2')
plt.xscale('log')
plt.xlabel('Amplitude')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

```python
inc1 = 30
dec1 = 30
rotatated_vector1 = ma.rotate_vector(r_vector1*amp1, inc1, dec1)

rotatated_vector1 = r_vector1*amp1


inc2 = -50
dec2 = -140
rotatated_vector2 = ma.rotate_vector(r_vector2*amp2, inc2, dec2)
```

## Generate grid coordinates

```python
sensor_sample_distance = 5.0  # µm
coordinates = vd.grid_coordinates(
    region=[0, 2000, 0, 1400],  # µm
    spacing=2,  # µm
    extra_coords=sensor_sample_distance,
)
```

```python
# bz_regional = mg.dipole_bz(coordinates, deep_dipole_coordinates, deep_dipole_moments)
bz = mg.dipole_bz(coordinates, dipole_coordinates, dipole_moments)
# data = vd.make_xarray_grid(coordinates, bz+bz_regional, data_names=["bz"], dims=("y", "x"), extra_coords_names="z")
data = vd.make_xarray_grid(coordinates, bz, data_names=["bz"], dims=("y", "x"), extra_coords_names="z")
data.x.attrs = {"units": "µm"}
data.y.attrs = {"units": "µm"}
data.bz.attrs = {"long_name": "vertical magnetic field", "units": "nT"}


# Add some random noise to make it more realistic.
noise_std_dev = 50  # nT
data.bz.values += mg.gaussian_noise(noise_std_dev, data.bz.shape, seed=42)
# Add a positive shift to the data 
data.bz.values += 200 # nT

data
```
