# Rotation of Directions Measured in the 2G Magnetometer  

The goal of this notebook is to adjust the direction measured by the 2G magnetometer by applying a -90Â° rotation around the X-axis. This correction ensures that the reference frames of the 2G magnetometer and the Field of Views (FoVs) of the QDM are aligned for consistent analysis.  


```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches
import pathlib

import sys
sys.path.append("..")
import micromag as mg
```

```python
# Function for rotation around the X-axis
def rotate_x(vector, theta):
    Rx = np.array([[1, 0, 0],
                   [0, np.cos(theta), -np.sin(theta)],
                   [0, np.sin(theta), np.cos(theta)]])
    return np.dot(Rx, vector)

# Function for rotation around the Y-axis
```

# Rotate Ceramic NRM direction

```python
dec, inc = 141.7, 18.8 # ceramic

vector = mg.angles_to_vector(inc, dec, 1)

# vector = rotate_z(vector, np.radians(90))
vector = rotate_x(vector, np.radians(-90))

projection = mg.StereographicProjection(np.array([vector]))
tmp = projection.plot()
```

```python
inc, dec, _ = mg.vector_to_angles(vector)
print(f" Declination: {dec:.2f}Â°\n Inclination: {inc:.2f}Â°")
```
